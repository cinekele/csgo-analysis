```python
for game_id in range(1, 334):
    log.info(f"Fixing grenades from game: {game_id}")
    tick = pd.read_sql(sql = f'SELECT tick FROM frame WHERE matchID = {game_id}', con = db_con)["tick"]
    grenade = pd.read_sql(sql = f'SELECT * FROM grenade WHERE matchID = {game_id}', con = db_con)
    throwTick_parsed = grenade["throwTick"].apply(lambda x: np.max(tick[tick<=x])).fillna(np.min(tick))
    destroyTick_parsed = grenade["destroyTick"].apply(lambda x: np.max(tick) if x == 0 else np.max(tick[tick<=x])).fillna(np.min(tick))
    for throwTick_p, destroyTick_p, throwTick, destroyTick in zip(throwTick_parsed, destroyTick_parsed, 
                                                                  grenade["throwTick"], grenade["destroyTick"]):
        with db_con.connect() as connection:
            connection.execute(f"UPDATE grenade SET throwTick_parsed = {throwTick_p}, destroyTick_parsed = {destroyTick_p}"
                               + f" WHERE matchID = {game_id} AND throwTick = {throwTick} AND destroyTick = {destroyTick}")
```
